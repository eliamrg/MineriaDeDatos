```python
import pandas as pd
import matplotlib.pyplot as plt
import os

#IMPORTACION DEL DATASET
data = pd.read_csv('dataSet\MoviesDataSet.csv')

df_by_genre = data.groupby(['title_year'])['imdb_score'].mean().reset_index()
df_by_genre.boxplot(by = 'title_year',figsize=(14,7))
plt.xticks(rotation=45)
plt.title('Calificacion promedio por Ao')
plt.xlabel('Anio')
plt.ylabel('Calificacion')
plt.show()
```

```python


df_by_genre = data.groupby(['title_year'])['gross'].sum().reset_index()
df_by_genre.boxplot(by = 'title_year',figsize=(14,7))
plt.xticks(rotation=45)
plt.title('Suma Ganancias por Ao (Miles de Millones DLS)')
plt.xlabel('Anio')
plt.ylabel('Ganancias')
plt.show()
```
