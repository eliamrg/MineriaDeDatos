```python
import pandas as pd
import matplotlib.pyplot as plt

# Importación del Dataset
data = pd.read_csv('dataSet\MoviesDataSet.csv')

# Agrupación por año y cálculo de la calificación promedio
df_by_year = data.groupby(['title_year'])['imdb_score'].mean().reset_index()

# Gráfico de líneas en lugar de un boxplot
plt.figure(figsize=(14, 7))
plt.plot(df_by_year['title_year'], df_by_year['imdb_score'], marker='o', linestyle='-', color='b')

# Ajustes de estilo y etiquetas
plt.xticks(rotation=45)
plt.title('Calificacion promedio por Año')
plt.xlabel('Año')
plt.ylabel('Calificacion Promedio')
plt.grid(True)
plt.show()


```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Importación del Dataset
data = pd.read_csv('dataSet\MoviesDataSet.csv')

# Agrupación por año y cálculo de la suma de ganancias
df_by_year = data.groupby(['title_year'])['gross'].sum().reset_index()

# Gráfico de líneas en lugar de un boxplot
plt.figure(figsize=(14, 7))
plt.plot(df_by_year['title_year'], df_by_year['gross'] / 1e9, marker='o', linestyle='-', color='b')

# Ajustes de estilo y etiquetas
plt.xticks(rotation=45)
plt.title('Suma Ganancias por Año (Miles de Millones DLS)')
plt.xlabel('Año')
plt.ylabel('Ganancias (Miles de Millones DLS)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importación del Dataset
data = pd.read_csv('dataSet\MoviesDataSet.csv')

# Agrupación por año y cálculo de la calificación promedio
df_by_year_score = data.groupby(['title_year'])['imdb_score'].mean().reset_index()

# Agrupación por año y cálculo de la suma de ganancias
df_by_year_gross = data.groupby(['title_year'])['gross'].sum().reset_index()

# Crear figura y subgráficos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 14), sharex=True)

# Gráfico de líneas para la calificación promedio
ax1.plot(df_by_year_score['title_year'], df_by_year_score['imdb_score'], marker='o', linestyle='-', color='b')
ax1.set_title('Calificacion promedio y Suma Ganancias por Año')
ax1.set_ylabel('Calificacion Promedio')
ax1.grid(True)

# Gráfico de líneas para la suma de ganancias
ax2.plot(df_by_year_gross['title_year'], df_by_year_gross['gross'] / 1e9, marker='o', linestyle='-', color='r')
ax2.set_xlabel('Año')
ax2.set_ylabel('Ganancias (Miles de Millones DLS)')
ax2.grid(True)

# Ajustes generales
plt.xticks(rotation=45)
plt.show()
```
