```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv('dataSet\MoviesDataSet.csv')
```

```python
import matplotlib.pyplot as plt

def get_cmap(n, name="hsv"):
    # Funcin para obtener un mapa de colores (cmap) con el nombre y nmero especificados
    return plt.cm.get_cmap(name, n)

def data_dispersion(labelx, labely, df):
    # Funcin para crear una grfica de dispersin a partir de un DataFrame

    # Obtiene los aos nicos presentes en la columna 'title_year'
    years = df['title_year'].unique()

    # Obtiene un mapa de colores con base en la cantidad de aos nicos ms uno
    cmap = get_cmap(len(years) + 1)

    # Crea una figura de Matplotlib con un tamao especfico
    plt.figure(figsize=(12, 6))

    # Itera sobre los aos y crea un grfico de dispersin para cada ao
    for i, year in enumerate(years):
        # Filtra el DataFrame para obtener un subconjunto de datos correspondiente al ao actual
        subset = df[df['title_year'] == year]

        # Crea un grfico de dispersin para el subconjunto actual
        plt.scatter(subset[labelx], subset[labely], label=year, color=cmap(i), alpha=0.5)

    # Muestra una leyenda con los aos
    plt.legend()

    # Establece el ttulo de la grfica
    plt.title('Relacion entre la calificacion en IMDB y las ganancias')

    # Etiqueta el eje x
    plt.xlabel('Calificacion en IMDB')

    # Etiqueta el eje y
    plt.ylabel('Ganancias')

    # Muestra una cuadrcula en la grfica
    plt.grid(True)

    # Muestra la grfica de dispersin
    plt.show()

# Llama a la funcin con las etiquetas 'imdb_score' y 'gross' para los ejes x e y, respectivamente, y el DataFrame 'data'
data_dispersion('imdb_score', 'gross', data)



# Generar una grfica de dispersin que representa la relacin entre la calificacin en IMDB y 
# las ganancias para cada ao en el conjunto de datos.
```
