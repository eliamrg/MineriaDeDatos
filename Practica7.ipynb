```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv('dataSet\MoviesDataSet.csv')
```

```python
import matplotlib.pyplot as plt

def get_cmap(n, name="hsv"):
    # Función para obtener un mapa de colores (cmap) con el nombre y número especificados
    return plt.cm.get_cmap(name, n)

def data_dispersion(labelx, labely, df):
    # Función para crear una gráfica de dispersión a partir de un DataFrame

    # Obtiene los años únicos presentes en la columna 'title_year'
    years = df['title_year'].unique()

    # Obtiene un mapa de colores con base en la cantidad de años únicos más uno
    cmap = get_cmap(len(years) + 1)

    # Crea una figura de Matplotlib con un tamaño específico
    plt.figure(figsize=(12, 6))

    # Itera sobre los años y crea un gráfico de dispersión para cada año
    for i, year in enumerate(years):
        # Filtra el DataFrame para obtener un subconjunto de datos correspondiente al año actual
        subset = df[df['title_year'] == year]

        # Crea un gráfico de dispersión para el subconjunto actual
        plt.scatter(subset[labelx], subset[labely], label=year, color=cmap(i), alpha=0.5)

    # Muestra una leyenda con los años
    plt.legend()

    # Establece el título de la gráfica
    plt.title('Relacion entre la calificacion en IMDB y las ganancias')

    # Etiqueta el eje x
    plt.xlabel('Calificacion en IMDB')

    # Etiqueta el eje y
    plt.ylabel('Ganancias')

    # Muestra una cuadrícula en la gráfica
    plt.grid(True)

    # Muestra la gráfica de dispersión
    plt.show()

# Llama a la función con las etiquetas 'imdb_score' y 'gross' para los ejes x e y, respectivamente, y el DataFrame 'data'
data_dispersion('imdb_score', 'gross', data)



# Generará una gráfica de dispersión que representa la relación entre la calificación en IMDB y 
# las ganancias para cada año en el conjunto de datos.
```
