```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el conjunto de datos desde el archivo CSV
data = pd.read_csv('dataSet\MoviesDataSet.csv')

# Filtrar los valores no nulos en la columna 'title_year'
years = data['title_year'].dropna()

# Convertir los aos a cadenas antes de contar la frecuencia
year_counts = years.astype(int).astype(str).value_counts()

# Comprobar si hay al menos un valor no nulo
if not year_counts.empty:
    # Crear la nube de palabras
    wordcloud = WordCloud(
        background_color="white",   # Color de fondo de la nube de palabras
        width=800,                   # Ancho de la imagen de la nube de palabras
        height=800,                  # Altura de la imagen de la nube de palabras
        max_font_size=100            # Tamao mximo de la fuente en la nube de palabras
    ).generate_from_frequencies(year_counts)

    # Configurar la visualizacin de la nube de palabras
    plt.figure(figsize=(8, 8), facecolor=None, dpi=300)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")  # Desactivar ejes
    plt.tight_layout(pad=0)

    # Guardar la nube de palabras como una imagen PNG
    plt.savefig('./word_cloud.png', format='png', bbox_inches='tight', pad_inches=0, dpi=300)

    # Mostrar la nube de palabras
    plt.show()
else:
    print("No hay valores no nulos en la columna 'title_year'.")

```
